```python
codebook_xls = pd.ExcelFile('UK Accidents - Codebook.xls')
district_descrip = pd.read_excel(codebook_xls,'Local Authority (District)')
```

```python
district_descrip.rename({'label': 'county'},axis=1,inplace=True)
```

```python
district_descrip
```

Xử lí với Scotland và Northern Ireland

```python
postcode_copy = postcode_df.copy()
```

```python
t1=postcode_copy[['county']].copy()
t1['code']=0
t1.set_index('county',inplace=True)
t1.update(district_descrip.set_index('county'))
t1.reset_index()
t1['code']=t1['code'].astype(int)
```

```python
t1
```

```python
postcode_copy = postcode_copy.merge(t1,on='county',how='left')
postcode_copy.rename({'code': 'Local_Authority_(District)'},axis=1,inplace=True)
postcode_copy
```

```python
df = postcode_copy[['Local_Authority_(District)','county','country_name','region_name']]
df
```

```python
df.drop_duplicates(subset='county',keep='first',inplace=True)
df
```

```python
df = df[df.country_name.isin(['Scotland','Northern Ireland'])]
df
```

```python
df.country_name.value_counts()
```

```python
df.to_csv('LDA_to_dim_Scotland_NIreland.csv')
```
